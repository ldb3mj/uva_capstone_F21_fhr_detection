## This code randomly samples from the NONE images folders to reduce the NONE samples.  

#### Remember to move all images not wanted from the /common/images/ subfolders that you do NOT want sampled for the neural network.  

#### Remove BRADY, TACHY and DEC_PROLONG up a level and out of the /images/ directory!

```python
#https://gist.github.com/spdin/9a693f2f0cd4195718423d24d308dc0a
import os, random, shutil

#Prompting user to enter number of files to select randomly along with directory
#specify directory file path to sample from
source='../images/NONE/'
dest='../None_images_removed/'
no_of_files=int(input("Enter The Number of Files To Select : "))   #4000

print("%"*25+"{ Details Of Transfer }"+"%"*25)
print("\n\nList of Files Moved to %s :-"%(dest))

#Using for loop to randomly choose multiple files
for i in range(no_of_files):
    #Variable random_file stores the name of the random file chosen
    random_file=random.choice(os.listdir(source))
    print("%d} %s"%(i+1,random_file))
    source_file="%s/%s"%(source,random_file)
    dest_file=dest
    #"shutil.move" function moves file from one directory to another
    shutil.move(source_file,dest_file)

print("\n\n"+"$"*33+"[ Files Moved Successfully ]"+"$"*33)
```

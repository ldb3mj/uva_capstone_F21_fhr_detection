```python
df1001 = pd.read_csv('./database/signals/1001.csv')
```

```python
df1001.tail(100)
```

```python
from matplotlib.pyplot import figure



df1001['samples'] = df1001.seconds / 4.0
df1001['FHR_int'] = df1001.FHR.replace(0,np.NaN)

df1001.FHR_int = df1001.FHR_int.interpolate(method='polynomial', order=5)

figure(figsize=(20, 6), dpi=80)
plot1 = plt.figure(1)
plt.plot(df1001.seconds.loc[:450], df1001.FHR.loc[:450])

figure(figsize=(20, 6), dpi=80)
plot2 = plt.figure(2)
plt.plot(df1001.seconds.loc[:450], df1001.FHR_int.loc[:450])

# plt.plot(df1001.seconds, df1001.UC)
plt.xlabel("time/samples")
plt.ylabel("FHRbpm")
plt.show()
```
